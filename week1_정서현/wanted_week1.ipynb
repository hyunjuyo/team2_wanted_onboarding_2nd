# 성장률로 유니콘기업 찾기


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


plt.rcParams['font.family'] = 'AppleGothic'
data = pd.read_csv("company_nps_data.csv", encoding ='euc-kr')
```

### 필요한 칼럼만 골라내기 

```python
data = data[["회사ID","연매출액","년도"]]
data
data = data.sort_values(by='년도')
data

data = data.drop_duplicates()
data
```

### str.contain사용을 위해 년도 str로 바꾸기

```python
data = data.astype({"년도":"str"})
```

### 연매출액이 1조가 넘는 기업만 골라내기 

```python
annual_sales = data.연매출액 >= 1000000000
annual_sales = data.loc[annual_sales,:]

annual_sales["회사ID"].unique()
```

### 연매출이 1조가 넘는 기업들만 데이터에서 뽑아내기

```python
data

data = data.astype({"회사ID":"str"})
result_uni = data[data["회사ID"].str.contains("381180|297874|126516|227415|294649|227414|126831|419977|420008|419945|419998|126606|63")]
```

```python
result_uni 
```

### 년도별로 칼럼 정리

```python
data_2015 = result_uni[result_uni["년도"].str.contains("2015")]
data_2015

data_2015.rename(columns = {'연매출액': '2015'}, inplace = True)
data_2015 = data_2015.drop(columns = ["년도"])

data_2015 = data_2015.set_index('회사ID')
data_2015
```

```python
data_2016 = result_uni[result_uni["년도"].str.contains("2016")]
data_2016

data_2016.rename(columns = {'연매출액': '2016'}, inplace = True)
data_2016 = data_2016.drop(columns = ["년도"])

data_2016 = data_2016.set_index('회사ID')

data_2016
```

```python
data_2017 = result_uni[result_uni["년도"].str.contains("2017")]
data_2017

data_2017.rename(columns = {'연매출액': '2017'}, inplace = True)
data_2017 = data_2017.drop(columns = ["년도"])

data_2017 = data_2017.set_index('회사ID')

data_2017
```

```python
data_2018 = result_uni[result_uni["년도"].str.contains("2018")]
data_2018

data_2018.rename(columns = {'연매출액': '2018'}, inplace = True)
data_2018 = data_2018.drop(columns = ["년도"])

data_2018 = data_2018.set_index('회사ID')

data_2018
```

```python
data_2019 = result_uni[result_uni["년도"].str.contains("2019")]
data_2019

data_2019.rename(columns = {'연매출액': '2019'}, inplace = True)
data_2019 = data_2019.drop(columns = ["년도"])

data_2019 = data_2019.set_index('회사ID')

data_2019
```

```python
result = data_2015.join([data_2016, data_2017, data_2018, data_2019])
result
```

### 합계 포함해서 전체 값을 보자

```python
cols = ['2015','2016','2017','2018','2019']
result['sum'] = result[cols].sum(axis=1)
```

```python
result
```

### 필요없는 데이터 정리

```python
sum_result = result.sort_values(by='sum', ascending=False).head(20)
sum_result
#전부 2019데이터 없으니 2019 그냥 drop 하고 
#결측치 있는것들 다 없애자 ..
```

```python
sum_result = sum_result.drop(columns = ["2019"])
```

```python
sum_result
```

```python
cleaned_result = sum_result.dropna()
```

```python
cleaned_result
```

### 2018년 기준으로 성장률 계산

```python


cleaned_result["성장률"]=(cleaned_result["2018"]-cleaned_result["2017"])/cleaned_result["2017"]
```

```python
cleaned_result
```

### 성장률 top 10 

```python
cleaned_result = cleaned_result.sort_values(by='성장률', ascending=False)
```
