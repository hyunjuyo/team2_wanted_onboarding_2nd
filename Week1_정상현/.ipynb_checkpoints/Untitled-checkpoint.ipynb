### 3가지 추가 변수 생성
    - 규정에서 처럼 1인당 연금보험료는 임금과 연동되는 수치이기 때문에, 직원당 임금을 다음과 같이 간접적으로 추정함
    
    - 월별_직원별_매출액 지표가 높으면 직원 1명이 제공하는 매출액이 높으므로, 흔히 말하는 '혁신'이 일어난 기업일 가능성이 높음
    
    - 월별_임금별_매출액은 회사가 지불하는 임금대비 효율(효율적인 임금설정의 파악)

```python
월별_직원별보험료=df1.월별_연금보험료/df1.월별_직원수
df1['임금']=((df1.월별_연금보험료/0.09)/df1.월별_직원수)


df1['월별_직원별_매출액']=df1.월별_매출액/df1.월별_직원수.copy()


df1['월별_임금별_매출액']=df1.월별_매출액/df1.임금.copy()

df1=df1.dropna()
```

```python
# 실제 유니콘 데이터
uni_data=pd.read_csv('유니콘기업매출액.csv',encoding='cp949')
gr_rt_3=uni_data.groupby('기업명')[['매출액', '평균연봉', '인원수']].pct_change()
gr_rt_3.columns=['매출gr','연봉gr','인원수gr']
uni_data=pd.concat([uni_data,gr_rt_3.fillna(0)],axis=1)
```

```python
uni_data['연봉별매출액']=uni_data.매출액/uni_data.평균연봉
uni_data['직원별매출액']=uni_data.매출액/uni_data.인원수
```

```python
uni_data
```

```python
sal_q1=uni_data.매출gr.quantile(0.25)
sal_q3=uni_data.매출gr.quantile(0.75)
inc_q1=uni_data.연봉gr.quantile(0.25)
inc_q3=uni_data.연봉gr.quantile(0.75)
pop_q1=uni_data.인원수gr.quantile(0.25)
pop_q3=uni_data.인원수gr.quantile(0.75)

# 이건 다음 단계에 사용할 예정임(월단위로 변환)
sal_p_inc_q1=uni_data.연봉별매출액.quantile(0.25)/12
sal_p_inc_q3=uni_data.연봉별매출액.quantile(0.75)/12
sal_p_pop_q1=uni_data.직원별매출액.quantile(0.25)/12
sal_p_pop_q3=uni_data.직원별매출액.quantile(0.75)/12
```

```python
df1_new=df1.groupby(['회사ID','year'])[['월별_매출액','월별_직원수','임금']].sum()
df1_gr=df1_new.groupby('회사ID')[['월별_매출액','월별_직원수','임금']].pct_change().dropna()
df1_gr.columns=['매출_gr','직원수_gr','임금_gr']
df1_gr
```

```python
# 실 유니콘 기업의 데이터에서 매출액,임금,직원수의 변동율의 하위 25%부터 상위75%까지의 범위 기반으로 연금 데이터를 필터링
df1_gr_1=df1_gr.loc[df1_gr.매출_gr.between(sal_q1,sal_q3),:]
df1_gr_1=df1_gr_1.loc[df1_gr_1.직원수_gr.between(pop_q1,pop_q3),:]
df1_gr_1=df1_gr_1.loc[df1_gr_1.임금_gr.between(inc_q1,inc_q3),:]
df1_gr_1
```
